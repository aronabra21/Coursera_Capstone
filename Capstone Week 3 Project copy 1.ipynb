```python
from pandas.io.html import read_html
page = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

wikitables = read_html(page, attrs={"class":"wikitable"})


wikitables[0].head()
wikitables[0].dropna(subset=["Neighborhood"], axis=0, inplace=True)
wikitables[0].shape and wikitables[0]
wikit=wikitables[0].groupby('Postal code')['Neighborhood'].apply(" , ".join).reset_index()
wikit1=wikit.drop('Postal code',axis=1)
wikit1
wikitable= wikitables[0].reset_index()
wikitable2=wikitable.sort_values(by=['Postal code'])
wikitable2=wikitable2.reset_index(drop=True)
wikitable3=wikitable2.drop(['index','Postal code'],axis=1)
wikitable3
```
